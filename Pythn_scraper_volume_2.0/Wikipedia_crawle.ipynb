```python
from bs4 import BeautifulSoup
import re
import random
import datetime
import requests

random.seed(datetime.datetime.now())
search_query = input("Enter the Search name: ")
search_query = search_query.title()

if " " in search_query:
    search_query = search_query.replace(" ","_")

def getLinks(articleURL):
    page = requests.get(f"https://en.wikipedia.org{articleURL}")
    soup = BeautifulSoup(page.content, "html.parser")
    return soup.find('div', {"id":"bodyContent"}).find_all('a', href= re.compile("^(/wiki/)((?!:).)*$"))

sub_links = getLinks(f"/wiki/{search_query}")

while len(sub_links):
    newArticles = sub_links[random.randint(0, len(sub_links)-1)].attrs['href']
    print(newArticles)
    sub_links = getLinks(newArticles)
```
